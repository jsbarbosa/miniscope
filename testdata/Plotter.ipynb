```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mergeChannel(channel, max = 4):
    datas = [np.genfromtxt("%stime%d.dat"%(channel, i),
                           delimiter = ",", skip_header=1) for i in range(1, max)]
    
    new = [max-1] + list(datas[0].shape)
    
    new = np.zeros(new)
    for i in range(max - 1):
        new[i] = datas[i]
        
    mean = new.mean(axis = 0)
    std = new.std(axis = 0)
    
    return mean, std
```

```python
def timeDiff(array):
    return 100 * abs(array[:, 1] - array[:, 2])/array[:, 1]
def speedDiff(array):
    return 100 * abs(array[:, 3] - array[:, 4])/array[:, 3]
```

```python
def change(means, name):
    fig, (ax1, ax2) = plt.subplots(2, sharex = True)
    y1 = timeDiff(meanz)
    ax1.plot(means[:, 0], y1, "-o")
    ax2.plot(means[:, 0], means[:, 3], "-o", alpha = 0.5, label = "Trace")
    ax2.plot(means[:, 0], means[:, 4], "-o", alpha = 0.7, label = "Retrace")
    
    ax1.set_ylabel("Variación (Trace vs Retrace) %")
    ax2.set_ylabel("Rapidez ($\mu$m/s)")
    ax2.set_xlabel("Resolución")
    
    ax2.legend()
    
    fig.tight_layout()
    
    fig.savefig(name)
    return fig
```

```python
meanz, stdz = mergeChannel("z")
meanx, stdx = mergeChannel("x", 3)
meany, stdy = mergeChannel("y", 2)
```

```python
change(meanz, "z.png")
```
